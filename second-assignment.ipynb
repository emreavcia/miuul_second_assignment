```python
import seaborn as sns
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = sns.load_dataset("titanic")
print(df)
```

```python
number_of_female_male = df["sex"].value_counts()
print(number_of_female_male)
```

```python
unique_values = [{"column_name": column_name, "num_unique_value": len(df[column_name].unique())} for column_name in df.columns]
print(unique_values)
```

```python
number_of_unique_values_pclass = df["pclass"].value_counts()
print(number_of_unique_values_pclass)
```

```python
number_of_unique_values_pclass = df["pclass"].nunique()
number_of_unique_values_parch = df["parch"].nunique()
print(f"pclass number of the uniques: {number_of_unique_values_pclass}, parch number of the uniques: {number_of_unique_values_parch}")
```

```python
print(df["embarked"].dtypes)
df.drop(columns = ["embarked"])
df["embarked"] = df["embarked"].astype("category")
print(df["embarked"].dtypes)
```

```python
print(df[df["embarked"] == "C"])
```

```python
print(df[df["embarked"] != "S"])
```

```python
print(df[(df["sex"] == "female") & (df["age"] < 30)])
```

```python
print(df[(df["fare"] > 500) | (df["age"] > 70)])
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.drop( columns="who", inplace=True)
print(df)
```

```python
deck_mode = df["deck"].mode()[0]
df["deck"] = df["deck"].fillna(deck_mode)
print(df["deck"].isnull().values.any())
```

```python
age_median = df["age"].median()
df["age"] = df["age"].fillna(age_median)
print(df["age"].isnull().values.any())
```

```python
survived_stats = df.groupby(["pclass", "sex"])["survived"].agg(["sum", "count", "mean"])
print(survived_stats)
```

```python
df["flag"] = df["age"].apply( lambda x: 1 if x < 30 else 0)
print(df["flag"])
```

```python
df2 = sns.load_dataset("tips")
print(df2)
```

```python
total_bill_by_time_variables = df2.groupby(["time"])["total_bill"].agg(["sum", "min", "max", "mean" ])
print(total_bill_by_time_variables)
```

```python
total_bill_by_time_and_day = df2.groupby(["time", "day"])["total_bill"].agg(["sum", "min", "max", "mean"])
print(total_bill_by_time_and_day)
```

```python
filtered_total_bill_by_female_and_lunch = df2[(df2["time"] == "Lunch") & (df2["sex"] == "Female")]
total_bill_by_female_and_lunch = filtered_total_bill_by_female_and_lunch.groupby(["day"])["total_bill"].agg(["sum", "min", "max", "mean"])
print(total_bill_by_female_and_lunch)
```

```python
filtered_df_by_size_and_total_bill = df2.loc[(df2["size"] < 3) & (df2["total_bill"] > 10)]
print(filtered_df_by_size_and_total_bill["total_bill"].mean())
```

```python
df2["total_bill_tip_sum"] = df2["total_bill"] + df2["tip"]
print(df2["total_bill_tip_sum"])
```

```python
sorted_by_total_bill_tip_sum = df2.sort_values("total_bill_tip_sum")
print(sorted_by_total_bill_tip_sum)
new_df = df2.loc[0:29]
print(new_df)
```
